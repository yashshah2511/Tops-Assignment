```python
# 1. Explain Supervised vs. Unsupervised Learning
# Supervised Learning

# The model is trained with data where the answer is already known.

# Example:
# Input: House size
# Output: House price


# The model learns the relationship and predicts future prices.

# Types of Supervised Learning

# Classification → Predict categories
# Example: Spam or Not Spam

# Regression → Predict continuous values
# Example: Salary prediction







# Unsupervised Learning

# The model works on data without labels and tries to discover patterns.

# Example:

# Group customers based on shopping behavior.

# Types of Unsupervised Learning

# Clustering → Group similar data
# Example: Customer segmentation

# Association → Find relationships
# Example: “People who buy milk also buy bread.”
```
